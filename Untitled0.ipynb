```python
!pip install circlify
import circlify
df1=df.groupby(['category'])['value'].sum().reset_index().sort_values(by='value', ascending=False).reset_index()
df11=df.groupby(['category'])['value'].sum().reset_index().sort_values(by='value', ascending=True).reset_index()
# compute circle positions:
circles = circlify.circlify(
    df1['value'].tolist(),
    show_enclosure=False,
    target_enclosure=circlify.Circle(x=0, y=0, r=4)
)

circles
```

```python
import matplotlib.pyplot as plt
# Find axis boundaries
fig, ax = plt.subplots(figsize=(6,6), layout='constrained')
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

for j,circle in enumerate(circles):
    x, y, r = circle
    c1='#7c4ea6'
    if(df11['category'][j]=='Bed'):
      c1='#f9c154'
    elif(df11['category'][j]=='Bath'):
      c1='#5094c3'
    circles1 = circlify.circlify(
    df[df['category']==df11['category'][j]]['value'].tolist(),
    show_enclosure=True,
    target_enclosure=circlify.Circle(x=x, y=y, r=r)
    )
    for i,circle1 in enumerate(circles1):
      x1, y1, r1 = circle1
      if(i==len(circles1)-1):
        ax.add_patch(plt.Circle((x1, y1), r1, ec=c1, fill=False))
      else:
        ax.add_patch(plt.Circle((x1, y1), r1, ec="#000000", fc=c1))
        ax.text(x, y, df11['category'][j], size=5, ha="center", color="w")
plt.axis("off")
```
